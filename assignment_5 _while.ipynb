```python
Guess a number is a game that prompts a player to guess a number between 0 and 9, 
which  is randomly generated by the system. 
the input given by the user matches the  number generated by the system then the user wins. 
The game should go as follows: 
Guess the number: 5 
Sorry, try again 
Guess the number: 3 
Sorry, try again 
Guess the number: 8 
You got it right! Congo! 
```

```python
import random
ran=random.randint(0,9)
while(True):
    
    x=eval(input("Enter a number between 0 to 9:"))

    if 0<=x<=9:

        if x==ran:
            print("You got it right ! Congo")
            break
        else:
            print("Sorry, try again")
            
    else:
        print("Please enter value between 0 to 9")
        
```

```python
2. Make an improvement to the Guess a number game. 
Guide the user where they are  standing and limit the number of attempts to 3. 
For example, the game should go like this: 
Guess the number: 4 
Too low 
Guess the number: 9 
Too high 
Guess the number: 5 
Sorry you Lost! 
```

```python
chances=eval(input("enter the number of chances")) # we can directly take this value to 3
i=0
ran=random.randint(0,9)

while(i<chances):
    x=eval(input("Enter a number between 0 to 9:"))
    
    diff=abs(ran-x)  

    if 0<=x<=9:

        if x==ran:
            print("You got it right ! Congo")
            break
            
        elif chances-1-i==0:
            
            print("Sorry you lost")
            
        else:
            if diff==1:
                print("Very close guess another number")
                
            elif diff>1:  
                print("The enter very value is too high")
                
            elif diff<1:
                print("Enter value is too low")
        
            print("You have only {} attempt left".format(chances-1-i))
            
    else:
        print("Please enter value between 0 to 9")
        print("You have only {} attempt left".format(chances-1-i))
    i+=1    
```

```python
#3. Let us make the above game a little more interesting by converting it 

into a gambling  problem. Suppose that a player starts with Rs. 1,000. 
a player can guess the number in his  first chance, 
then he will be given a prize of Rs. 5,000, if he requires 2 attempts then he will  get a prize of Rs.1,000. 
If he loses then he will lose Rs. 500. For example the game should go  like this: 
You have a cash of Rs. 1,000 with you...  
Guess the number: 8 
Too high 
Guess the number: 3 
You have just won Rs. 1,000
Your balance: Rs. 2000 
```
